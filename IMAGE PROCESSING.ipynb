# Color and Edge Analysis of Fabric
*A python based image-processing mini project*

## Introduction

This project emphasis on the color and edge analysis of a fabric sample using Python's OpenCv and MatplotLib Libraries. It is to examine the pixel-level color distribution and structural features of the tartan fabric swatch, which would be beneficial to textile sector of quality control, defect detection, and visual fabric classification.

The process includes- 
- Loading and Displaying fabric image in RGB form.
- Converting the image to Grayscale for structural features.
- Using of sharpening filter for enhancing thread and weave visibility.
- Evaluating RGB intensity levels of the image by using color histogram.
- Detecting edges by Canny method within the fabric.

---

## Step-1: Import the necessary libraries 
* Importing OpenCV  for image processing
* Matplotlib for graph and histogram
* NumPy for math array operation


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Step-2: Load and Display Fabric image

The fabric image is imported in BGR format (OpenCV's default). It is then converted to RGB from BGR to accurately display the orginal image color distribution with Matplotlib.

```python
# Load the image as path
image_path = 'Tartan_fabric.png'
image = cv2.imread(image_path)

# check if the image loaded properly
if image is None:
    print("Error! Image not found. Check your file path.")
else:
    image = cv2.resize(image, (600, 600)) # resize the image
    print("Image loaded and resized successfully!")
```

```python
# Show the imported image
plt.imshow(image)
plt.title("Orginal Image - BGR", fontweight = 'bold') # visible in BGR
plt.axis('off')
plt.show() 
```

```python
# Convert to RGB from BGR
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Show the image in RGB
plt.imshow(image_rgb)
plt.title('Orginal Image - RGB', fontweight = 'bold')
plt.axis('off')
plt.show() # imported in original color
```

## Step-2: Grayscale and Sharpening</u>🪄
- The image is converted to GrayScale version to acquire clearer contrast for structural examination.
- Convolution filter is applied using a sharpening kernel to improve the visibility of fabric threads and weave structure by enhancing details and edge clarity.


```python
# Convert to Grayscale
gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Sharpening Filter
kernel_sharpened = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
sharpened = cv2.filter2D(image_rgb, -1, kernel_sharpened)

print("GrayScale and Sharpened version created")
```

```python
# Show the GrayScale and sharpened version
 
# Canvas size
plt.figure(figsize=(10,5))

# GrayScale Display
plt.subplot(1,2,1) # one row, two column and placed in index one
plt.imshow(gray, cmap='gray')
plt.title('GrayScale', fontweight = 'bold')
plt.axis('off')


# Sharpened Display
plt.subplot(1,2,2) # one row, two column and placed in index two
plt.imshow(sharpened)
plt.title('Sharpened', fontweight = 'bold')
plt.axis('off')

plt.tight_layout() # adjust spacing and fits clean and nicely
plt.show()
```

## Step-3: Color Analysis - RGB Histogram🪄

The histogram shows how much red or green or blue is present throughout the fabric image. This is useful for fabric dye inspection, color balancing or understanding dominant color tones.

**Note:** This raw histogram shows the exact distribution of `pixel intensity` values without applying any post-pocessing to allow raw analysis of color distribution.

```python
# Color Histogram for RGB channels

colors = ('r', 'g', 'b')
labels = ('Red Channel', 'Green Channel', 'Blue Channel')
plt.figure(figsize = (12,6))

for i, col in enumerate(colors):
    hist = cv2.calcHist([image_rgb], [i], None, [256], [0,256])
    plt.plot(hist, color = col, linewidth = 2, label = labels[i] )
    plt.xlim([0,256])

plt.title('RGB Color Distribution',fontsize = 16, fontweight= 'bold')
plt.xlabel('Pixel Intensity', fontsize = 12)
plt.ylabel('Frequency', fontsize = 12)
plt.grid(alpha = 0.3)
plt.legend(fontsize = 12)
plt.tight_layout()
plt.show()



```

## Step-4: Edge Detection - Canny Filter🪄

Edge detection helps us in finding edges, weave and potential defects which is necessary for analyzing fabric texture, structure and overall quality of the fabric.

**Threesholds used:** 50 (lower) , 150 (upper)

```python

edges = cv2.Canny(gray, 50, 150)

plt.figure(figsize = (10,5))
plt.imshow(edges, cmap = 'gray')
plt.title('Edge Detection - Canny', fontsize= 16, fontweight= 'bold')
plt.axis('off')
plt.tight_layout()
plt.show()
```
